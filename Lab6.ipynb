```
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

#AD 1 
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
```

```
#AD 2
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=343)
```

```
#AD 3
clf1 = SVC(C=0.5, kernel='linear')
clf1.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf1, legend=2)
y_pred = clf1.predict(x_test)
plt.xlabel('Long [cm]')
plt.ylabel('Width [cm]')
plt.title('C=0.5 and kernel=linear')
plt.show()
print(f'Precision:{metrics.accuracy_score(y_test, y_pred)}')
```

```
#AD 4
clf2 = SVC(C=0.05, kernel='poly')
clf2.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf2, legend=2)
y_pred2 = clf2.predict(x_test)
plt.xlabel('Long [cm]')
plt.ylabel('Width [cm]')
plt.title('C=0.05 and kernel=poly')
plt.show()
print(f'Precision:{metrics.accuracy_score(y_test, y_pred2)}')

clf3 = SVC(C=0.2137, kernel='rbf')
clf3.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf3, legend=2)
y_pred2 = clf3.predict(x_test)
plt.xlabel('Long [cm]')
plt.ylabel('Width [cm]')
plt.title('C=0.2137 and kernel=rbf')
plt.show()
print(f'Precision:{metrics.accuracy_score(y_test, y_pred2)}')

clf4 = SVC(C=0.7, kernel='sigmoid')
clf4.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=clf4)
y_pred2 = clf4.predict(x_test)
plt.xlabel('Long [cm]')
plt.ylabel('Width [cm]')
plt.title('C=0.7 and kernel=sigmoid')
plt.show()
print(f'Precision:{metrics.accuracy_score(y_test, y_pred2)}')
```
